```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
mnist = tf.keras.datasets.mnist  #importing datasets

(xx_train,y_train),(x_test,y_test) = mnist.load_data() #loading them in train and test

x_train= tf.keras.utils.normalize(xx_train, axis=1) # perprocessing normalizing the values of tensor to 1 as max

x_test= tf.keras.utils.normalize(x_test, axis=1) # perprocessing normalizing the values of tensor to 1 as max
```

```python
model= tf.keras.models.Sequential() # importing a Sequential model
model.add(tf.keras.layers.Flatten()) # used to flatten 
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu)) # fully connected layers
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu)) # fully connected layers
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu)) # fully connected layers
model.add(tf.keras.layers.Dense(10,activation=tf.nn.sigmoid)) # output layer

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
model.fit(x_train,y_train,epochs=10)
```

```python
val_loss, val_acc=model.evaluate(x_test,y_test) # for model valdation
```

```python
model.save('holylight.model') # to save the model so that we dont need to run this code again and again
```
